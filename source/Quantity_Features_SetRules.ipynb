```python
from temporal_relation.parse.ParseTimebankPT import TimebankPT
import warnings
warnings.filterwarnings('ignore')
# warnings.simplefilter('ignore')


path_tml = r'temporal_relation\TimeBankPT\*\*.tml'
tb = TimebankPT(path_tml, dev=False, ignore_load_corpus=False)  # %time
print(tb)

# Carrega features
tb.tr.setRules_start('dataset/dataset_features')
tb.features.load_dataset('dataset/dataset_features.parquet')
```

```python
# CBA
print('\nCBA')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_CBA.txt', 'my_rules')
tb.tr.setRules.remove_rule_class_default()
cba = tb.tr.setRules.copy()
print('TOTAL REGRAS CBA:', len(cba.rules))

# CN2
print('\nCN2')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_CN2.txt', 'my_rules')
tb.tr.setRules.remove_rule_class_default()
cn2 = tb.tr.setRules.copy()
print('TOTAL REGRAS CN2:', len(cn2.rules))

# IDS
print('\nIDS')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_IDS.txt', 'my_rules')
tb.tr.setRules.remove_rule_class_default()
ids = tb.tr.setRules.copy()
print('TOTAL REGRAS IDS:', len(ids.rules))

# RIPPER
print('\nRIPPER')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_RIPPER.txt', 'my_rules')
tb.tr.setRules.remove_rule_class_default()
ripper = tb.tr.setRules.copy()
print('TOTAL REGRAS RIPPER:', len(ripper.rules))


# TODOS SETRULES
print('\nTODOS SETRULES')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_Combinados_Todos.txt', 'my_rules')
tb.tr.setRules.remove_rule('origem', 'MANUAL')
tb.tr.setRules.remove_rule_class_default()
todos_setrules = tb.tr.setRules.copy()
print('TOTAL TODOS SETRULES:', len(todos_setrules.rules))


# TODOS SETRULES
print('\nTODOS SETRULES')
tb.tr.setRules.clear()
tb.tr.load_rules('rules/FINAL_Combinados_Todos.txt')
tb.tr.remove_rule_class_default()
df_todos_tr = tb.tr.df_rules.copy()
print('TOTAL TODOS SETRULES:', len(df_todos_tr))


# COMBINADO DE 2 EM 2
print('\nCOMBINADO DE 2 EM 2')
tb.tr.setRules.clear()
tb.tr.setRules.add_setRules('rules/FINAL_Combinados_2_em_2.txt', 'my_rules')
tb.tr.setRules.remove_rule_class_default()
setrules_2_em_2 = tb.tr.setRules.copy()
print('TOTAL SETRULES 2 EM 2:', len(setrules_2_em_2.rules))
```

```python
import pyperclip

def print_resumo(setRules, quant_features: int = 10, verbosy = True):
    
    list_features = []

    df = setRules.df_features_counts
    size = df.shape[0]
    if verbosy: print('Features utilizadas:', size)
    
    if type(quant_features) not in [int, float]:
        quant_features = size
    
    if verbosy: print('\nFEATURES MAIS FREQUENTES:')
    for i, feat in df.sort_values('Total', ascending=False).head(quant_features)[['Total']].iterrows():
        list_features.append(i[0].replace("_", "-") + '=' + str(int(feat['Total'])))
    
    return list_features


def print_quant_regras(setRules):
    quant_regras = len(setRules.rules)
    quant_predicado = setRules.df_features_counts['Total'].sum()
    print('\nREGRAS:', quant_regras, '\nPREDICADOS:', quant_predicado,
          '\nMÉDIA:', quant_predicado / quant_regras)


features_52 = ['event_between_order', 'reichenbach_direct_modification', 'event_class', 'event_closest_to_event_pos', 'event_closest_to_timex3_pos', 'event_dep', 'timex3_dep', 'event_pos_token_1_precede', 'event_pos_token_1_follow', 'event_pos_token_2_precede', 'event_pos_token_2_follow', 'event_pos_token_3_precede', 'event_pos_token_3_follow', 'timex3_pos_token_1_follow', 'timex3_pos_token_2_precede', 'timex3_pos_token_2_follow', 'timex3_pos_token_3_precede', 'timex3_pos_token_3_follow', 'event_preposition_precede', 'timex3_preposition_precede', 'event_timex3_distance', 'timex3_relevant_lemmas', 'event_gov_verb_tense', 'timex3_gov_verb_tense', 'timex3_head_pos', 'signal_precede_timex3_text', 'signal_precede_timex3_pos', 'reichenbach_tense',
               'timex3_preposition_gov', 'timex3_pos', 'event_closest_to_event_class', 'event_temporal_direction', 'event_modal_verb', 'event_closest_to_event_temporal_direction', 'timex3_type', 'timex3_between_order', 'event_conjunction_closest_precede', 'event_conjunction_closest_follow', 'signal_precede_event_child_event', 'event_closest_to_event_equal_class', 'event_closest_to_event_equal_pos', 'event_closest_to_event_equal_tense', 'event_closest_to_timex3_equal_pos', 'timex3_pos_token_1_precede', 'signal_precede_event_text', 'event_closest_to_event_tense', 'reichenbach_temporal_mod_function', 'event_intervening_preceding_class', 'event_preposition_gov', 'timex3_is_ancestor_event', 'signal_precede_timex3_dep_if_child_timex3', 'event_polarity']
features_41 = ['event_closest_to_event_equal_class', 'event_between_order', 'reichenbach_direct_modification', 'reichenbach_temporal_mod_function', 'event_class', 'event_closest_to_event_class', 'event_closest_to_event_pos', 'event_closest_to_event_tense', 'event_closest_to_timex3_pos', 'event_conjunction_closest_follow', 'event_dep', 'timex3_dep', 'timex3_type', 'timex3_pos', 'event_pos_token_1_precede', 'event_pos_token_1_follow', 'event_pos_token_2_precede', 'event_pos_token_2_follow', 'event_pos_token_3_precede', 'event_pos_token_3_follow', 'timex3_pos_token_1_precede',
               'timex3_pos_token_1_follow', 'timex3_pos_token_2_precede', 'timex3_pos_token_2_follow', 'timex3_pos_token_3_precede', 'timex3_pos_token_3_follow', 'event_preposition_precede', 'timex3_preposition_precede', 'event_timex3_distance', 'event_closest_to_event_temporal_direction', 'event_temporal_direction', 'timex3_relevant_lemmas', 'event_gov_verb_tense', 'timex3_gov_verb_tense', 'timex3_head_pos', 'event_preposition_gov', 'signal_precede_timex3_text', 'signal_precede_timex3_pos', 'signal_precede_event_child_event', 'reichenbach_tense', 'event_modal_verb']
features_70 = tb.features.X_train.columns
```

## CBA

```python
setrules = cba
print_quant_regras(setrules)
print('\n', setrules.df_rules.value_counts('relType'), '\n')

pyperclip.copy(str(print_resumo(setrules, 'max', verbosy = False)))
for feat in print_resumo(setrules):
    print(feat)

display(setrules.df_features_counts.head(10),
        setrules.df_features_counts.tail(5))
print('\nFEATURES NÃO UTILIZADA:')
print(set(features_41) -
      set(setrules.df_features_counts.reset_index()['feature'].unique()))
```
